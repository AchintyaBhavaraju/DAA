```python
import time
from random import randint
import matplotlib.pyplot as plt
import sys

sys.setrecursionlimit(3000)


def merge(arr, l, m, r):
    n1 = m - l + 1
    n2 = r - m

    L = [0] * (n1)
    R = [0] * n2

    for i in range(0, n1):
        L[i] = arr[l + i]

    for j in range(0, n2):
        R[j] = arr[m + 1 + j]

    i = 0 
    j = 0 
    k = l  

    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1


    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1

    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1

def mergeSort(arr, l, r):
    if l < r:

        m = l + (r - l) // 2
        mergeSort(arr, l, m)
        mergeSort(arr, m + 1, r)
        merge(arr, l, m, r)

arr = [12, 11, 13, 5, 6, 7]
n = len(arr)
print("Given array is")
for i in range(n):
    print("%d" % arr[i], end=" ")

mergeSort(arr, 0, n - 1)
print("\n\nSorted array is")
for i in range(n):
    print("%d" % arr[i], end=" ")

arr_1 = [randint(0, 3000) for i in range(0, 5000)]
size_1 = len(arr_1)

def plot(x_list, times):
    plt.xlabel("No. of elements")
    plt.ylabel("Time required")
    plt.plot(x_list, times)
    plt.show()


list1 = [i for i in range(500)]

times = []
for x in range(0, 10000, 10):
    start_time = time.time()
    mergeSort(arr_1, 0, size_1-1)
    elapsed_time = time.time() - start_time
    times.append(elapsed_time)
print(times)
x_1 = [i for i in range(0, 10000, 10)]
plot(x_1, times)
```

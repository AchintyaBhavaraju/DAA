```python
import time
from random import randint
import matplotlib.pyplot as plt
import sys

def findMinAndMax(nums, left, right, min=sys.maxsize, max=-sys.maxsize):
    if left == right:
        if min > nums[right]: 
            min = nums[right]
        if max < nums[left]: 
            max = nums[left]
        return min, max

    if right - left == 1:

        if nums[left] < nums[right]:
            if min > nums[left]:
                min = nums[left]
            if max < nums[right]:
                max = nums[right]
        else:
            if min > nums[right]: 
                min = nums[right]
            if max < nums[left]:  
                max = nums[left]
        return min, max

    mid = (left + right) // 2


    min, max = findMinAndMax(nums, left, mid, min, max)
    min, max = findMinAndMax(nums, mid + 1, right, min, max)
    return min, max


nums = [7, 2, 9, 3, 1, 6, 7, 8, 4]

(min, max) = findMinAndMax(nums, 0, len(nums) - 1)

print("The minimum element in the list is", min)
print("The maximum element in the list is", max)


arr_1 = [randint(0, 5000) for i in range(0, 5000)]
size_1 = len(arr_1)

def plot(x_list, times):
    plt.xlabel("No. of elements")
    plt.ylabel("Time required")
    plt.plot(x_list, times)
    plt.show()

list1 = [i for i in range(500)]
times = []
for x in range(0, 10000, 10):
    start_time = time.time()
    findMinAndMax(arr_1, 0, size_1-1)
    elapsed_time = time.time() - start_time
    times.append(elapsed_time)
print(times)
x_1 = [i for i in range(0, 10000, 10)]
plot(x_1, times)
```

```python
#BINARY SEARCH 
import time
from random import randint
import matplotlib.pyplot as plt


def binary_search(arr, low, high, x):
    if high >= low:

        mid = (high + low) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] > x:
            return binary_search(arr, low, mid - 1, x)
        else:
            return binary_search(arr, mid + 1, high, x)

    else:
        return -1


# Test array
arr = [2, 3, 4, 10, 40]
# x = 10
x = randint(0, 1000)
# Function call
result = binary_search(arr, 0, len(arr) - 1, x)

if result != -1:
    print("Element is present at index", str(result))
else:
    print("Element is not present in array")


def plot(x_list, times):
    plt.xlabel("No. of elements")
    plt.ylabel("Time required")
    plt.plot(x_list, times)
    plt.show()

list1 = [i for i in range(500)]

x1 = randint(0, 500)
times = []
for x in range(0, 100000, 10):
    start_time = time.time()
    binary_search(list1, 0, len(arr) - 1, x)
    elapsed_time = time.time() - start_time
    times.append(elapsed_time)
print(times)
x_1 = [i for i in range(0, 100000, 10)]
plot(x_1, times)
```

```python
#LINEAR SEARCH
import time
from random import randint
import matplotlib.pyplot as plt


def linearSearch(array, n, x):
 
    for i in range(0, n):
        if (array[i] == x):
            return i
    return -1


# Test array
arr = [2, 3, 4, 10, 40]
# x = 10
x = randint(0, 1000)
# Function call
result = linearSearch(arr, len(arr) - 1, x)

if result != -1:
    print("Element is present at index")
else:
    print("Element is not present in array")


# ------------Plotting Graph--------#
def plot(x_list, times):
    plt.xlabel("No. of elements")
    plt.ylabel("Time required")
    plt.plot(x_list, times)
    plt.show()


# ---------------Worst Case Time Complexity--------#
list1 = [i for i in range(500)]

x1 = randint(0, 500)
times = []
for x in range(0, 100000, 10):
    start_time = time.time()
    linearSearch(list1,len(arr) - 1, x)
    elapsed_time = time.time() - start_time
    times.append(elapsed_time)
print(times)
x_1 = [i for i in range(0, 100000, 10)]
plot(x_1, times)
```

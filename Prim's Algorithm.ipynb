```python
import matplotlib.pyplot as plt
from datetime import datetime
import math

# Get Array 
import random
import time
no_of_arrays = int((input('Please enter number of graphs you need: ')))
list_number_of_elements_in_array = []
list_of_list_of_edges = []

graphs = []
p = 0.9
for i in range(no_of_arrays):
    number_of_nodes = int(input("Please enter number of nodes in a graph: "))
    list_of_list_of_weights = []
    for j in range(number_of_nodes):
        list_of_weights= [random.randint(20,100) for i in range(number_of_nodes)]
        list_of_list_of_weights.append(list_of_weights)
    graphs.append(list_of_list_of_weights)
print(len(graphs))

import sys

class Graph():
 
    def __init__(self, vertices):
        self.V = vertices
        self.graph = [[0 for column in range(vertices)]for row in range(vertices)]
   
    def printMST(self, parent):
        print("Edge \tWeight")
        for i in range(1, self.V):
            print(parent[i], "-", i, "\t", self.graph[i][parent[i]])
    
    def minKey(self, key, mstSet):
        min = sys.maxsize
        for v in range(self.V):
            if key[v] < min and mstSet[v] == False:
                min = key[v]
                min_index = v
        return min_index
 
    def primMST(self):
        key = [sys.maxsize] * self.V
        parent = [None] * self.V  
        key[0] = 0
        mstSet = [False] * self.V
        parent[0] = -1 
 
        for cout in range(self.V):
            u = self.minKey(key, mstSet)
            mstSet[u] = True
            for v in range(self.V):
                if self.graph[u][v] > 0 and mstSet[v] == False and key[v] > self.graph[u][v]:
                    key[v] = self.graph[u][v]
                    parent[v] = u
        self.printMST(parent)

import matplotlib.pyplot as plt
time_taken = []
inputs = []
for graph in graphs:
    start = time.perf_counter()
    g = Graph(len(graph))
    g.graph = graph
    g.primMST()
    end = time.perf_counter()
    t = end - start
    inputs.append(len(list(graph)))
    time_taken.append(t)
plt.plot(inputs, time_taken, marker = 'o')
plt.title("Prims")
plt.xlabel("Number of Edges in a graph")
plt.ylabel("time_taken")

if __name__ == '__main__':
    g = Graph(5)
    g.graph = [[0, 3, 1, 6, 8],
               [3, 0, 2, 8, 4],
               [1, 4, 0, 7, 6],
               [6, 9, 0, 0, 8],
               [0, 5, 7, 9, 0]]
 
    g.primMST()
 
```
